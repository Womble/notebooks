```python
spectra=[]
for k in merged :
    if not(k=='1450' or 'cut' in k):
        RT=merged[k].createFreeFreeRT((300,300,300))
        spectra.append(RT.spectrum(logspace(log10(2e9),log10(30e9))))
        %xdel RT
        gc.collect()
```

```python
leg=[]
for i,k in enumerate([x for x in merged if not(x=='1450' or 'cut' in x)]) :
    loglog(logspace(log10(2e9),log10(30e9)), spectra[i])
    leg.append(float(k)/10)
legend(leg, loc='best')
title('face-on spectra of models')
gca().set_xlim(1e9,4e10)
gca().set_xlabel('frequency / Hz')
gca().set_ylabel('flux / mJy')
savefig('faceSpectra.png', bbox='tight')
savefig('faceSpectra.eps', bbox='tight')
```
