```python
ims0={}
ims45={}
ims90={}
for i in merged:
    RT=merged[i].createFreeFreeRT((250,250,200))
    ims0[i]=RT.rayTrace(5e9);   
    ims90[i]=RT.rayTrace(5e9, transpose=1);
    RT.rotatecube(45, trim=1)
    gc.collect()
    ims45[i]=RT.rayTrace(5e9);   
    gc.collect()
```

```python
plot(*zip(*[(float(i)/10.0, ims0[i].sum()) for i in ['55','77','100','117','162']])+['+'])
title('face on')
```
