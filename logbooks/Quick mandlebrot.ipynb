```python
gd=mgrid[-1.5:0:1024j,-0.75:0.75:1024j]
c=empty((128,128),dtype=complex)
c=gd[0,...]+(0+1j)*gd[1,...]
z=zeros_like(c)

counter=zeros_like(z, dtype=int)
```

```python
limit=5
def do_mandy(z,counter):
    z[...]=0
    counter*=0
    for i in xrange(1,250):
        z*=z
        z+=c
        z[z!=z]=+inf
        counter[abs(z)<limit]+=1
```
