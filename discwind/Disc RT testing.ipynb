```python
rotvels=[]
for k in ks:
  if not('cut' in k):
    d=merged[k]
    mask=d.discMaterial(centralMass=10*dap.m_sol)
    mask*=np.logical_not(d.boundGas(centralMass=10*dap.m_sol))
    norm=(d.rho()*d.ionisation())**2*mask*d.vol
    rotvels.append((d.Uphi()*norm)[:,100:].sum()/norm[:,100:].sum())
figure()
plot([merged[k].unitLength/dap.r_sol for k in ks if not ('cut' in k)],rotvels,'+')
gca().set_xlabel(r'r/r_\odot')
gca().set_ylabel(r'r_phi')
gca().set_xlim(4,17)
```
