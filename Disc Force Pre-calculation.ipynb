```python
#problem constants
L_star=8.5e3*L_sol
r_star=7.7*r_sol   
m_star=10*m_sol

filename='disc_1e-3_07p7'

MMM=(0.909*1.008*const.atomic_mass+0.088*4.002602*const.atomic_mass+0.003*12*const.atomic_mass)
cs0=(sqrt(const.Boltzmann*13.8e3/MMM))

def cs (R):
    return cs0*pow(R/r_star*sqrt(r_star/r_sol/5.5), 0.5 * -2.0/4.0)
def H (R):
    return cs(R)/sqrt(m_star*const.G/R**3)

Hprime=ut.funcDiff(H)
ThetaHprime=ut.memoize(lambda R : math.atan(Hprime(R)))
```

```python
def disc_lum (M_star, acc_rate, r_star):
    return const.gravitational_constant*M_star*acc_rate/(2*r_star)
print '%.4e W'%disc_lum(10*m_sol,1e-4*m_sol/year,5.5*r_sol) #5.5r_sol 10m_sol protostar accreating at 1e-4 solar masses per year
print '%.4e'%(m_sol*const.gravitational_constant/2/r_star*1e-14*m_sol)
```

```python
def disc_GeomOLD (r, M_star, L_star, acc_rate, r_star):
    rsr=abs(r_star/r)
    Ld=disc_lum(M_star,acc_rate,r_star)
    x=L_star/Ld
    try:
        return rsr**3*(1-math.sqrt(rsr)) + x/(3*pi) * (math.asin(rsr)-rsr*math.sqrt(1-rsr**2))
    except ValueError:
        print 'error',rsr
```

```python
def disc_Geom (r, M_star, L_star, acc_rate, r_star):
    rsr=abs(r_star/r)
    Ld=disc_lum(M_star,acc_rate,r_star)
    x=L_star/Ld
    theta=math.asin(rsr)
    try:
        return rsr**3*(1-math.sqrt(rsr)) + x/(3*pi) * (theta-rsr*math.cos(theta+ThetaHprime(r)))
    except ValueError:
        print 'error',rsr
```
